# Yelp Data Quality EDA process




```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, when
from pyspark.sql.types import StringType
from datetime import datetime

# Initialize Spark Session
spark = SparkSession.builder.appName("Yelp_EDA_Visualizations").getOrCreate()

def get_parquet_size(path):
    try:
        hadoop_conf = spark._jsc.hadoopConfiguration()
        path_obj = spark._jvm.org.apache.hadoop.fs.Path(path)
        fs = path_obj.getFileSystem(hadoop_conf)
        size_bytes = fs.getContentSummary(path_obj).getLength()
        return size_bytes
    except Exception as e:
        print(f"⚠️ Warning: Unable to retrieve size for {path}. {e}")
        return 0

def get_readable_size(size_bytes):
    return f"{size_bytes/(1024*1024):.2f} MB"

def compute_missing_counts(df):
    """
    Computes missing values per column for a DataFrame.
    For string columns, counts both NULL and empty string ("").
    For non-string columns, counts only NULL.
    Returns a dictionary mapping column name to missing count.
    """
    missing_expr = []
    for field in df.schema.fields:
        c = field.name
        if isinstance(field.dataType, StringType):
            missing_expr.append(sum(when((col(c).isNull()) | (col(c) == ""), 1).otherwise(0)).alias(c))
        else:
            missing_expr.append(sum(when(col(c).isNull(), 1).otherwise(0)).alias(c))
    missing_counts = df.select(missing_expr).collect()[0].asDict()
    return missing_counts

# Define the tables and their corresponding Parquet paths
tables = {
    "business": "/data/yelp/parquet/business",
    "checkin": "/data/yelp/parquet/checkin",
    "review": "/data/yelp/parquet/review",
    "user": "/data/yelp/parquet/user",
    "tip": "/data/yelp/parquet/tip"
}

for tname, path in tables.items():
    try:
        print(f"\n===== EDA for Table: {tname} =====")
        # Load the data from Parquet
        df = spark.read.parquet(path)
        
        # Compute total records and missing counts
        total_records = df.count()
        missing_counts = compute_missing_counts(df)
        file_size_bytes = get_parquet_size(path)
        file_size = get_readable_size(file_size_bytes)
        
        print(f"Total Records: {total_records}")
        print("Missing Counts per Column:")
        for col_name, cnt in missing_counts.items():
            print(f"  {col_name}: {cnt}")
        
        # Convert missing counts to a Pandas DataFrame for visualization
        missing_pd = pd.DataFrame(list(missing_counts.items()), columns=["Column", "MissingCount"])
        
        # Create a bar plot for missing counts
        plt.figure(figsize=(12, 6))
        sns.barplot(x="Column", y="MissingCount", data=missing_pd, palette="viridis")
        plt.title(f"Missing Values Count for {tname.capitalize()} Table\n(Total Records: {total_records}, File Size: {file_size})")
        plt.xlabel("Column")
        plt.ylabel("Missing Values")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()
        
        # (Optional) Show summary statistics using describe() on numeric columns
        print("Summary Statistics (Numeric Columns):")
        df.select([col(c) for c in df.columns if dict(df.dtypes)[c] in ["int", "bigint", "double", "float"]]).describe().show(truncate=False)
        
    except Exception as e:
        print(f"❌ EDA for {tname} failed: {str(e)}")

print("\n🎉 Data Quality EDA process completed for all tables!")
```

# **📖 TLDR Understanding Yelp Dataset with Storytelling**

This analysis aims to **tell the story** behind Yelp’s ecosystem by breaking down the dataset into five key components:

1. **Business Listings** - Understanding what businesses are listed, where they are located, and their characteristics.
2. **Reviews** - Exploring customer sentiment, ratings, and how they shape business success.
3. **Users** - Identifying patterns in user activity, loyalty, and engagement.
4. **Check-ins** - Investigating how frequently businesses are visited.
5. **Tips** - Analyzing short but useful recommendations shared by users.
---

## **🔍 Chapter 1: The Yelp Business Landscape**
The **Business dataset** gives us a **comprehensive overview** of the businesses listed on Yelp. Each entry represents a business with details such as:
- **Location (City, State)**
- **Categories (e.g., Restaurant, Shopping, Health & Medical)**
- **Operational hours**
- **Star Ratings (1-5)**
- **Total number of reviews received**



### **Analysis:**
We will analyze the distribution of businesses across different cities and states using the `business` table. Additionally, we will explore the most common business categories and their average ratings. Correlation analysis will be conducted between `review_count` and `stars` to see if highly reviewed businesses tend to have higher ratings.

---

```
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, avg

# Initialize Spark Session
spark = SparkSession.builder.appName("YelpAnalysis").getOrCreate()

# Load Business Data
business_df = spark.read.parquet("/data/yelp/parquet/business").toPandas()

# Business distribution by state
plt.figure(figsize=(12, 6))
sns.barplot(y=business_df['state'].value_counts().index, x=business_df['state'].value_counts().values)
plt.title("Business Distribution by State")
plt.xlabel("Count")
plt.ylabel("State")
plt.show()

# Average star rating by category
avg_rating = business_df.groupby("categories")["stars"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rating.values, y=avg_rating.index)
plt.title("Top 10 Categories by Average Star Rating")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.show()
```

## **👥 Chapter 3: Yelp Users – Who’s Engaging?**
The **User dataset** provides rich insights into **how people interact** with Yelp. Each user has:
- **Review Count**
- **Friends**
- **Elite Status (loyalty over years)**
- **Useful, Funny, and Cool Votes**

### **Analysis:**

```
user_df = spark.read.parquet("/data/yelp/parquet/user").toPandas()

# Distribution of review counts per user
plt.figure(figsize=(12, 6))
sns.histplot(user_df['review_count'], bins=50, kde=True)
plt.title("Distribution of Review Counts per User")
plt.xlabel("Review Count")
plt.ylabel("Number of Users")
plt.show()
```

## **📍 Chapter 4: Check-ins – Foot Traffic Trends**
The **Check-in dataset** tracks **how often customers visit businesses**. While reviews capture sentiment, check-ins tell us about **real-world engagement**.


```
# Load Check-in Data
checkin_df = spark.read.parquet("/data/yelp/parquet/checkin").toPandas()

# Businesses with the highest number of check-ins
top_checkins = checkin_df['business_id'].value_counts().head(10)
top_checkins.plot(kind='bar', figsize=(12, 6), title="Top Businesses by Check-ins")
plt.xlabel("Business ID")
plt.ylabel("Check-ins")
plt.show()
```

## **💡 Chapter 5: The Power of Tips**
The **Tip dataset** provides **short-form recommendations**, often **focusing on a single feature** of a business. Unlike reviews, which are **long and detailed**, tips are quick insights like:
- "Try the spicy tuna roll!"
- "Best place for brunch in town."
- "Avoid weekends – too crowded!"

```
# Load Tip Data
tip_df = spark.read.parquet("/data/yelp/parquet/tip").toPandas()

# Businesses with the most tips
top_tips = tip_df['business_id'].value_counts().head(10)
top_tips.plot(kind='bar', figsize=(12, 6), title="Top Businesses by Tips")
plt.xlabel("Business ID")
plt.ylabel("Number of Tips")
plt.show()
```
